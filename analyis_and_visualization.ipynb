```python
import pandas as pd

df = pd.read_csv('/content/LinkedIn_Jobs_python_developer_toronto_2025-02-25.csv')
df.head()
```

```python
top_job_titles =df['Title'].value_counts().head(10)
print(top_job_titles)
```

```python
import matplotlib.pyplot as plt

  # Plot top job titles
top_job_titles.plot(kind='bar', title='Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Number of Postings')
plt.show()
```

```python
  # Plot top locations
df['Date'] = pd.to_datetime(df['Date'])
postings_over_time = df.groupby(df['Date'].dt.to_period('M')).size()

  # Plotting
postings_over_time.plot(kind='line', title='Job Postings Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Postings')
plt.show()
```

```python
  # Scatter plot of applicants vs. posting date
plt.scatter(df['Date'], df['Applicants'])
plt.title('Applicants Over Time')
plt.xlabel('Date Posted')
plt.ylabel('Number of Applicants')
plt.show()
```
